# -> Subplot of 'Ram', 'OS', 'Touchscreen', 'IPSpanel', 'RetinaDisplay', 'CPU_company

```python
# Define the categorical columns to plot
columns_to_plot = ['Ram', 'OS', 'Touchscreen', 'IPSpanel', 'RetinaDisplay', 'CPU_company']

# Set up subplots
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(15, 10))  # Adjust rows and columns as needed
axes = axes.flatten()  # Flatten for easy iteration

# Create countplots for each categorical column
for i, col in enumerate(columns_to_plot):
    sns.countplot(x=df[col], ax=axes[i], palette="Set2")
    axes[i].set_title(f"Count of {col}")
    axes[i].set_ylabel("Count")
    axes[i].set_xticklabels(axes[i].get_xticklabels(), rotation=45)

# Adjust layout
plt.tight_layout()
plt.show()
```

### The most common RAM size is 8GB, followed by 4GB and 16GB, while higher capacities like 32GB and 64GB are rare.
### Windows 10 dominates the OS category, with very few devices running other operating systems.
### Touchscreen laptops are significantly fewer, making up a small portion of the total.
### Intel CPUs are overwhelmingly dominant, with AMD and Samsung having a minimal presence.
### Most devices lack IPS panels and Retina displays, indicating a preference for standard screen technology.
